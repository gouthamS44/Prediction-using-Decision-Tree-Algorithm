It applies a straitforward idea to solve the classification problem. 

Decision Tree Classifier poses a series of carefully crafted questions about the attributes of the test record. Each time time it receive an answer, a follow-up question is asked until a conclusion about the calss label of the record is reached.The purpose of this is to find when we feed any new data to this classifier, it would be able to

```python
#Now,Let's Build the Model.
x = iris_data
y = dat
from sklearn .model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state = 42)
#returns (number of row, number of column) for each training and test data.
print("Shape of feature training data :",x_train.shape)
print("Shape of target training data :",y_train.shape)
print("Shape of feature test data :",x_test.shape)
print("Shape of target test data :",y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score,confusion_matrix

clf = DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
# Finding accuracy for training data
predict = clf.predict(x_train)
print("Accuracy of training data : ",accuracy_score(predict,y_train)*100,"%")
print("Confusin matrix of training data :'\n' ",confusion_matrix(predict,y_train))
sns.heatmap(confusion_matrix(predict,y_train),annot = True,cmap = 'Greens')
```

```python
plt.figure(figsize  = (18,18))
tree.plot_tree(clf,filled = True,rounded = True,proportion = True,node_ids = True , feature_names = data.feature_names)
plt.show()
```
